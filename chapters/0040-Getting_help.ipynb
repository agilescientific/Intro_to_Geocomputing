## Getting help

The following are four basic built-in functions of Python that we can use to inspect how modules, libraries, and functions work.

- the `dir()` function

- the `help()` function

- the `type()` function

- the `__doc__` function

** `dir()` function **

The `dir` function is one of the most useful functions in Python when it comes to digging around in Python's interactive mode. You provide it with the name of a module or object, and it returns a list of the contents of that module or object. 

```python
dir(math)
```

** help() function **

Call this function on a module, object, function, method, or attribute and it will return the help documentation about it.

```python
dir(10)
```

```python
help(math)
```

```python
help(str)
```

** type() function**

Call this function on an object and Python will return the object's type. Type checking can come in handy when you are debugging.

```python
type(math)
```

```python
type(math.cos)
```

```python
type(math.cos(10))
```

** `__doc__` function **

Python documentation strings (or docstrings) provide a convenient was of associating textual description with modules, functions, classes, and methods. The docstring in any Python object is specified by the programmer including a string constant ast he first statement in the object's definition. Docstrings describe what the module, object, or function does, and how it should be used. All functions should be written with a docstring at the start other users and our future selves can inspect these comments as an interactive help system.

```python
math.__doc__
```
